## Bloch-Sphere Representation of a Qubit  
A general Qubit can be expressed as $\ket{\psi} = \alpha_0\ket{0} + \alpha_1\ket{1}$ with $\lvert \alpha_0\rvert^2 + \lvert \alpha_1\rvert^2 = 1$. Anothe important aspect to mention is that in Quantum Computing two state vectors which only differ by a unitary complex factor -in other words, by a so-called global phase- are considered equivalent. In symbols, a state vector $\ket{\psi}$ is considered equivalent to $e^{i\theta}\ket{\psi}$, with $\theta\in\mathbb{R}$ -technically, they represent the same equivalence class.  
Performing some elementary algebraic manipulations, one can arrive at the following general represention of a Qubit (ignoring global phases): $$\ket{\psi} = cos(\theta/2)\ket{0} + e^{i\phi}sin(\theta/2)\ket{1}$$
with $\theta, \phi \in \mathbb{R}$. If we now define the Bloch-Vector of $\psi$ to be $n_{\theta, \phi}: = [sin(\theta)cos(\phi), sin(\theta)sin(\phi), cos(\theta)]^T \in \mathbb{R}^3$, we can see that $n_{\theta, \phi}$ is an element of the surface of the unit sphere in $\mathbb{R}^3$, where $\theta$, $\phi$ represent respectivelly the polar and azimuthal angles in polar coordinates. That is, there is a bijection between the Qubits and the surface of the unit sphere -which is called Bloch-Sphere in Quantum computing.

### Bloch-Sphere representation of $\ket{0} = cos(0)\ket{0} + e^{i0}sin(0/2)\ket{1}$:

```python
circuit = QuantumCircuit(1)
#plot_bloch_multivector(circuit, title="Bloch-Sphere Representation of |0>", figsize=(2,2))
plot_bloch_multivector(circuit, figsize=(2,2))
```

### Bloch-Sphere representation of $\ket{1} = cos(\pi/2/2)\ket{0} + e^{i0}sin(\pi/2)\ket{1}$:

```python
state_vector = Statevector.from_label('1')
circuit.initialize([0, 1], 0, 1)
#plot_bloch_multivector(circuit, title="Bloch-Sphere Representation of |->", figsize=(2,2))
plot_bloch_multivector(circuit, figsize=(2,2))
```

### Bloch-Sphere representation of $\ket{+} = cos\left(\dfrac{\pi/2}{2}\right)\ket{0} + e^{i0}sin\left(\dfrac{\pi/2}{2}\right)\ket{1}$:

```python
state_vector = Statevector.from_label('1')
circuit.initialize([1/sqrt(2), 1/sqrt(2)], 0, 1)
#plot_bloch_multivector(circuit, title="Bloch-Sphere Representation of |+>", figsize=(2,2))
plot_bloch_multivector(circuit, figsize=(2,2))
```

### Bloch-Sphere representation of $\ket{-} = cos\left(\dfrac{\pi/2}{2}\right)\ket{0} + e^{i\pi}sin\left(\dfrac{\pi/2}{2}\right)\ket{1}$:

```python
state_vector = Statevector.from_label('1')
circuit.initialize([1/sqrt(2), -1/sqrt(2)], 0, 1)
#plot_bloch_multivector(circuit ,title="Bloch-Sphere Representation of |->", figsize=(2,2))
plot_bloch_multivector(circuit, figsize=(2,2))
```
